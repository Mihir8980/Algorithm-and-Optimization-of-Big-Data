# Q 4A. Incremental Mathematical Stream Regression

## Generator Function 
### Generate_X for Independent variables
### Generate_y for Dependent variables

```python
def generate_X(X): #Generator function for sending data in the multiple of 100 (i.e 1-100,101-200,...) 
    for i in range(4): # Here i have consider 500 data point
        yield X[i*100:(i+1)*100,:]
def generate_y(y): #Generator function for sending data in the multiple of 100 (i.e 1-100,101-200,...)
    for i in range(4): # Here i have consider 500 data point
        yield y[i*100:(i+1)*100]
```

## IMSR Works on this equation :
![](m.PNG)

```python
# Importing Required Libraries
import numpy as np
from sklearn.datasets import load_boston
from numpy.linalg import pinv,inv,LinAlgError
from sklearn import metrics

# Hyperparameter
temp_XTX = np.zeros([14,14])
temp_XTy = np.zeros([1,14])
count=0
theta_list = []

# Loading Boston Dataset
X,y = load_boston(return_X_y=True)
dummy1 = np.ones((X[400:506].shape[0],1)) 
X_test = np.c_[dummy1,X[400:506]] # (106,13) data points for training
y_test = y[400:506] # 106 data points for training

# Creating instance of X and y
A = generate_X(X)
B = generate_y(y)    
    
# try-Catch block because to utilize whole data i have used .next of generator 
try:
    while True:
        X_train_temp1 = next(A) # For next X_train
        dummy = np.ones((X_train_temp1.shape[0],1)) 
        X_train = np.c_[dummy,X_train_temp1] # Appending one column in the front
        y_train = next(B) # For next y_train
        theta = np.zeros((X.shape[1],1)) # Initializing the parameter for current iteration

        temp_XTX = temp_XTX + np.dot(X_train.T,X_train)  # Synobsis for XTX
        try: # Inverse XTX
            XTXi = inv(temp_XTX)
        except:
            XTXi = pinv(temp_XTX)

        temp_XTy = temp_XTy + np.dot(X_train.T,y_train) # Synobsis for XTy
        theta = np.dot(XTXi,temp_XTy.T) # Computed theta
        theta_list.append(theta) # Theta list which contain all parameter
        print(10*"-")
        prediction = np.dot(X_test,theta) # theta_list[19] theta estimated based in 1-400 data points 
        print("MAE: ",metrics.mean_absolute_error(y_true=y_test,y_pred=prediction))
        print("MSE: ",metrics.mean_squared_error(y_true=y_test,y_pred=prediction))
        print(10*"-")
except Exception as e:
    print("Train Data Set is consumed")
```
